```python
# Import Libraries
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn
from nltk import pos_tag

# Define Dataset Location
folderPath = 'projectFiles'
```

```python
# Define Dataset Location
folderPath = 'projectFiles'

# Define Output File Path
output_path = os.path.join(folderPath, "wordnet_target_word_ambiguity.txt")

# Define Target Words
targetWords = {"soul", "love", "know", "visions", "skies"}
```

```python
# POS tag converter
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wn.ADJ
    elif treebank_tag.startswith('V'):
        return wn.VERB
    elif treebank_tag.startswith('N'):
        return wn.NOUN
    elif treebank_tag.startswith('R'):
        return wn.ADV
    else:
        return None

#Analyze files and print output to terminal and a Text File
with open(output_path, 'w', encoding='utf-8') as out:
    for filename in os.listdir(folderPath):
        if filename.endswith(".txt"):
            with open(os.path.join(folderPath, filename), 'r', encoding='utf-8') as f:
                text = f.read()

            tokens = word_tokenize(text.lower())
            tagged = pos_tag(tokens)

            header = f"\n--- {filename} ---\n"
            print(header.strip())
            out.write(header)

            for target in targetWords:
                count = tokens.count(target)

                tag = next((pos for word, pos in tagged if word == target), None)
                wn_pos = get_wordnet_pos(tag) if tag else None
                syns = wn.synsets(target, pos=wn_pos) if wn_pos else wn.synsets(target)

                line = f"{target} - Frequency: {count}, Synset meanings (POS: {wn_pos if wn_pos else 'any'}): {len(syns)}\n"
                print(line.strip())
                out.write(line)
```
